```python
import numpy as np
import matplotlib.pyplot as plt

umaze = np.load("/home/jaewoo/practices/diffusers/Augmentation-For-OfflineRL/data/maze2d-umaze-v1.npy", allow_pickle=True)


def merge_dictionary(list_of_Dict):
    merged_data = {}

    for d in list_of_Dict:
        for k, v in d.items():
            if k not in merged_data.keys():
                merged_data[k] = [v]
            else:
                merged_data[k].append(v)

    for k, v in merged_data.items():
        merged_data[k] = np.concatenate(merged_data[k])

    return merged_data

plt.plot(umaze[0]['rewards'])
```

```python
umaze = merge_dictionary(umaze)
```
