```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set font globally to Times New Roman
plt.rcParams['font.family'] = 'Times New Roman'
plt.rcParams['font.size'] = 10  # You can adjust the size if needed

def draw(df, name):
    plt.pcolor(df, cmap='magma')
    plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns)
    plt.yticks(np.arange(0.5, len(df.index), 1), df.index)
    plt.title(f'{name}', fontsize=20)
    plt.xlabel('μ', fontsize=14)
    plt.ylabel('α', fontsize=14)
    plt.colorbar()
    plt.show()


# Data definition
novelty = {
    "10": [19.642, 19.663, 19.684],
    "25": [19.669, 19.723, 19.746],
    "50": [20.01, 20.544, 21.243],
    "75": [20.028, 20.648, 21.378],
    "100": [20.028, 20.631, 21.373]
}

# Create a pandas DataFrame
novelty = pd.DataFrame(novelty)
novelty.index = [1.1, 1.2, 1.3]
novelty.sort_index(ascending=True)
# Set the size of the figur

data = {
    "10": [34.951, 34.957, 34.957],
    "25": [34.956, 35.002, 35.112],
    "50": [35.485, 36.335, 37.444],
    "75": [35.544, 36.416, 37.264],
    "100": [35.507, 36.392, 37.432]
}

# Creating the DataFrame
diversity = pd.DataFrame(data, index=[1.1, 1.2, 1.3])

data = {
    "10": [34.957, 0.002],
    "25": [35.112, 0.364],
    "50": [37.444, 10.356],
    "75": [37.264, 34.686],
    "100": [37.432, 34.918]
}

# Row labels
index = ["diversity", "L2 distance"]

# Creating the DataFrame
l2_diversity = pd.DataFrame(data, index=index)
```
