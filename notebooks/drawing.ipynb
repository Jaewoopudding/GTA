```python
import random
import numpy as np
import plotly.express as px
import matplotlib.pyplot as plt

def simulate_1d_rw(nsteps=1000, p=0.5, stepsize=1):
    steps = [ 1*stepsize if random.random() < p else -1*stepsize for i in range(nsteps) ]
    y = np.cumsum(steps)
    x = list(range(len(y)))

    return x, list(y)

simulation_data = {}
nsims = 5
for i in range(nsims):
    x, y = simulate_1d_rw()
    simulation_data['x'] = x
    simulation_data['y{col}'.format(col=i)] = y

ycols = [ 'y{col}'.format(col=i) for i in range(nsims) ]
fig = px.line(simulation_data, x='x', y=ycols)
fig.show()
```

```python
total_traj_len = 2000
```

```python
np.repeat([500, 1000, 1500, 2000], 2)
```

```python
def padding(arr):
    zeros = np.zeros(total_traj_len - arr.shape[0])
    return np.concatenate([zeros, arr])

def create_random_walk(len, dim):
    rws = []
    for d in range(dim):
        rws.append(padding(np.random.randn(len).cumsum()).reshape(1, -1))
    return np.concatenate(rws, 0)
```

```python
create_random_walk(1000, 1).shape
```

```python
plt.figure()
ax=plt.axes(projection='3d')
ratios = [500, 1000, 1500, 2000]
cmap = plt.get_cmap('winter')
for ratio in np.repeat(ratios, 3):
    x, y = create_random_walk(ratio, 2)
    z = range(total_traj_len)
    ax.plot3D(y, z, x, color=cmap(ratio/total_traj_len), alpha=0.5)
```
