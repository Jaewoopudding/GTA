```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python

norm = torch.distributions.Normal(loc=0, scale=0.0003)
```

```python
norm.sample((3,3))
```

```python
sample_num=100000
plt.hist(norm.sample_n(sample_num), bins=100, alpha=0.3)
plt.hist(np.random.randn(sample_num) * 0.0003, bins=100, alpha=0.3)
plt.show()
```
